A survey conducted by the Times of India shows the GDP of the country and the literacy rate of the country. Now using agglomerative algorithm cluster the countries which are having a similar GDP and literacy rate.

```python
import pandas as pd

data = pd.read_csv('./data/countries.csv')

data.head().T
```

We only need `country`, `literacy`, and `gdp`.

```python
cols = ['GDP ($ per capita)', 'Literacy (%)', 'Country']
unwanted_cols = [col for col in list(data.keys()) if col not in cols]
data = data.drop(unwanted_cols, axis=1)
```

```python
data.rename({'GDP ($ per capita)': 'gdp', 'Literacy (%)': 'literacy', 'Country': 'country'}, axis=1, inplace=True)
data.head()
```

```python
data['literacy'] = [float(str(literacy).replace(',', '.')) for literacy in data['literacy']]
data.iloc[:, 1:].head()
```

```python
from scipy.cluster.hierarchy import dendrogram

def plot_dendrogram(model, **kwargs):

    # Children of hierarchical clustering
    children = model.children_

    # Distances between each pair of children
    # Since we don't have this information, we can use a uniform one for plotting
    distance = np.arange(children.shape[0])

    # The number of observations contained in each cluster level
    no_of_observations = np.arange(2, children.shape[0]+2)

    # Create linkage matrix and then plot the dendrogram
    linkage_matrix = np.column_stack([children, distance, no_of_observations]).astype(float)

    # Plot the corresponding dendrogram
    dendrogram(linkage_matrix, **kwargs)
```

```python
from sklearn.cluster import AgglomerativeClustering

data = data.dropna()

model = AgglomerativeClustering()
model.fit(data.iloc[:, 1:].values)
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [16, 9]

plt.title('Hierarchical Clustering Dendrogram')
plot_dendrogram(model, labels=model.labels_)
plt.show()
```
