```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

### Traingular MF

```python
# New Antecedent/Consequent objects hold universe variables and membership
# functions
quality = ctrl.Consequent(np.arange(0, 11, 1), 'quality')
service = ctrl.Consequent(np.arange(0, 11, 1), 'service')
tip = ctrl.Consequent(np.arange(0, 26, 1), 'tip')

# Auto-membership function population is possible with .automf(3, 5, or 7)
quality.automf(3)
service.automf(3)

# Custom membership functions can be built interactively with a familiar,
# Pythonic API
tip['poor'] = fuzz.trimf(tip.universe, [0, 0, 13])
tip['average'] = fuzz.trimf(tip.universe, [0, 13, 25])
tip['good'] = fuzz.trimf(tip.universe, [13, 25, 25])
quality.view()
service.view()
tip.view()
```

### Z-Function MF

```python
quality['poor'] = fuzz.zmf(quality.universe,  0, 0)
quality['average'] = fuzz.zmf(quality.universe, 0, 5)
quality['good'] = fuzz.zmf(quality.universe, 5, 10)
service['poor'] = fuzz.zmf(service.universe,  0, 0)
service['average'] = fuzz.zmf(service.universe, 0, 5)
service['good'] = fuzz.zmf(service.universe, 5, 10)
tip['poor'] = fuzz.zmf(tip.universe,  0, 0)
tip['average'] = fuzz.zmf(tip.universe, 0, 13)
tip['good'] = fuzz.zmf(tip.universe, 13, 25)
quality.view()
service.view()
tip.view()
```

### Trapezoidal MF

```python
quality['poor'] = fuzz.trapmf(quality.universe,  [0, 0, 5, 10])
quality['average'] = fuzz.trapmf(quality.universe, [0, 5, 10, 10])
quality['good'] = fuzz.trapmf(quality.universe, [5, 10, 10, 15])
service['poor'] = fuzz.trapmf(service.universe,  [0, 0, 5, 10])
service['average'] = fuzz.trapmf(service.universe, [5, 10, 10, 15])
service['good'] = fuzz.trapmf(service.universe, [0, 0, 13, 25])
tip['poor'] = fuzz.trapmf(tip.universe, [0, 0, 13, 25])
tip['average'] = fuzz.trapmf(tip.universe, [0, 13, 25, 25])
tip['good'] = fuzz.trapmf(tip.universe, [13, 25, 25, 37])
quality.view()
service.view()
tip.view()
```
